```python
#  Jointplot

#  Jointplot creates a multi-panel figure that shows both the bivariate( or joint ) relationship between two
# variables between two variables along with the univariate margin (or marginal) distribution
#  of each on separate axes.

We can use Jointplots in the following 4 manners:
    1.ScatterPlots
    2.Hexbin Plots
    3.KdePlots
    4.RegPlot
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)   #plot individual hist on both the axis
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)

# higher for both values..it becomes darker
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)

# contor .. 2d .. PDF ..darker region heigher point on peak
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data=tips)

#  created aline which is known as best fit line.
# Draw a line segment
```

```python
Distplots are used to do univariate analysis on a continuous/numerical data

Distplot combines 3 things in one graph.
1.Histogram
2.Kernel Density Estimation(PDF)
3.Rugplot
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True)
```

```python
titanic['Cabin'].fillna(titanic['Cabin']=='Not specified',inplace=True)
```

```python
titanic.head()
```

```python
#  Plotting a Distplot

sns.displot(titanic['Age'],kde=True)
```

```python
# Only histogram

sns.distplot(titanic['Age'],bins=100,kde=False)
```

```python
# Plotting only the KDE
sns.distplot(titanic['Age'],hist=False)
```

```python
#  Plotting only the Rugplot
```

```python
sns.distplot(titanic['Age'],hist=False,kde=False,rug=True)

#  in a given range...more points tilted towards '60'.
#  small vert line for each observation..1 line-1 observation.
#  plot  the points side by side
```

```python
#  How to use the Displot
sns.distplot(titanic[titanic['Survived']==1]['Age'])

#  age values of passangers who survived the accident.
sns.distplot(titanic[titanic['Survived']==0]['Age'])

#  age values of passangers who survived the died.
#  p(survive>died).People saved their kids.

# Blue:
survive
# red:
Died
Child,women:
p(survive>dying)



# 
```
