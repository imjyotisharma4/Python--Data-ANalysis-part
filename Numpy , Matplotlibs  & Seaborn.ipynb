```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
print(x)
plt.plot(x)
```

```python
x=np.arange(1,10)
y=[x**2 for x in x]         #list comprehension traditional way of writing for loop  / new method
print(x,y)
plt.plot(x,y)
```

```python
# Single line Graph

x=np.arange(1,10)             #Same
plt.plot(x,[x**2 for x in x])
```

```python
# Multi line Graph

x=np.arange(1,10)
print(x,y)
plt.plot(x,[x**2 for x in x])
plt.plot(x,[x**3 for x in x])
plt.plot(x,[x**4 for x in x])
```

```python
# grid() - It displays the grid. By default it is True. If given false will not display

x=np.arange(5,6,0.1)                  #y=x, value of y correspond to x
print(x)
plt.plot(x)
plt.grid(True)
```

```python
# xlabel - It will display label of x axis
# ylabel - It will display label of y axis

plt.plot(x,[x**2 for x in x])
plt.grid(True)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
```

```python
# title - It displays the title of the graph

plt.plot(x,[x**3 for x in x])
plt.grid(True)
plt.title("This is my first graph")
```

```python
# axis - It displays the axis. We can also provide our own axis.

plt.plot(x,[x*25 for x in x])
plt.title("This is my first graph ")
plt.axis([4.5,6.5,100,150])                      # starting twom values -min of x axis, max of x axis--
                                                 #--last two values , min of y axis, max of y axis
```

```python
# Legend()  - Displays label of inside lines graph

x=np.arange(1,10)

plt.plot(x,[x**2 for x in x],label='y=x^2')
plt.plot(x,[x**3 for x in x],label='y=x^3')
plt.plot(x,[x**4 for x in x],label='y=x^4')
plt.legend()
```

```python
x=np.arange(0.0,6.0,0.01)

plt.plot(x,[x**2 for x in x],label='Square')
plt.plot(x,[x**3 for x in x],label='Cube')
plt.plot(x,[x**4 for x in x],label='4th Power')
plt.legend(loc=0)                            #By default-- Top-Left corner which corresponds to Zero.
plt.legend(loc="center")
```

```python
#Customized
#Line,Colour & Marker
#Colour

x=np.arange(0.0,6.0,0.01)

plt.plot(x,[x**2 for x in x],'r',label='Square')
plt.plot(x,[x**3 for x in x],'y',label='Cube')
plt.plot(x,[x**4 for x in x],'m',label='4th Power')
plt.legend(loc=0)                                         #By default-- Top-Left corner which corresponds to Zero.
plt.legend(loc="center")
```

```python
x=np.arange(0,6,1)

plt.plot(x,[x**2 for x in x],'r--',label='Square')
plt.plot(x,[x**3 for x in x],'g-D',label='Cube')
plt.plot(x,[x**4 for x in x],'m:',label='4th Power')
plt.legend(loc=0)                            #By default-- Top-Left corner which corresponds to Zero.
plt.legend(loc="center")
```

```python
#Color, line and marker

x=np.arange(1,25)

plt.plot(x,[x*2 for x in x],'r-s',label='Square')
plt.plot(x,[x*3 for x in x],'y-D',label='Cube')
plt.plot(x,[x*4 for x in x],'m:+',label='4th Power')
plt.legend(loc=0)                                            #By default  -- Top-Left corner which corresponds to Zero.
plt.legend(loc="center")
```

```python
import seaborn as sns
iris=sns.load_dataset('iris')                                      #in-built dataset
```

```python
iris.head()
```

```python
iris.tail()
```

```python
#Distplot -- Distribution Plot -Continuous(Quantity)
sns.distplot(iris["petal_length"])
```

```python
#Distplot -- Distribution Plot -Continuous(Quantity)
sns.distplot(iris["petal_length"],kde=False)

# kde, by default it is true.
```

```python
# Histogram --- needs continuous column

sns.histplot(iris["sepal_length"])
```

```python
# Scatter plot 

sns.scatterplot(iris['sepal_width'],iris['petal_width'])
```

```python
# Scatter plot -- - dependent variable(in vs with continuous data).....hue---broken by category wise...to break our scatter plot

sns.scatterplot(iris['sepal_width'],iris['petal_width'],hue=iris['species'])
```

```python
# Line plot -- mainly use on time series(continuous) analysis-quater/monthwise sales, trends, patterns

sns.lineplot(iris.index,iris['sepal_length'])  #X-axis 'iris.index'  & on y-axis,'sepal_length'


```

```python
# Barplot - Categorical column analysis.

sns.countplot(x=iris['species'])
```

```python
# Box plot - to detect the outliers(far from rest of the datapoints)- needs continuous columns

sns.boxplot(iris['sepal_width'])

```

```python
# Outliers can be detected on continuous data only

sns.boxplot(x=iris['species'],y=iris['sepal_width'])       #Categorical vs continuous analysis using boxplot
```

```python
# Pairplot - less numerical columns --- gives histogram & Scatter plot

sns.pairplot(iris,hue='species')

```

```python
# Pairplot - less numerical columns --- gives histogram & Scatter plot

sns.pairplot(iris)
```

```python
# Correlation -- 
corrDf=iris.corr()
corrDf.head()
```

```python
#For correlation Graph, we use Heatmap
sns.heatmap(corrDf)
```

```python
sns.heatmap(iris.corr(),annot=True)            # '+1'strongly positive , '-ve' strongly negtive

# annot is used to give values in graph
```
